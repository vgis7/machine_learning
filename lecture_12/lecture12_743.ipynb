```python
import numpy as np
import pandas as pd
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.svm import SVC
```

# Load the data:

```python
train_data = pd.read_csv("mnist_train.csv")
test_data = pd.read_csv("mnist_test.csv")

train_labels = train_data['label']
train_data.drop(columns=['label'])

test_labels = test_data['label']
test_data.drop(columns=['label'])
```

# Training the model with default estimator function:

```python
AdaBoost = AdaBoostClassifier(n_estimators=100)
AdaBoost.fit(train_data,train_labels)
```

```python
AdaBoost.score(test_data,test_labels)
```

# Traing the model with SVC as estimator

```python
svc=SVC(probability=True, kernel='linear')
```

```python
AdaBoostSVC = AdaBoostClassifier(n_estimators=100,base_estimator=svc)
AdaBoostSVC.fit(train_data,train_labels)
```

```python
AdaBoostSVC.score(test_data,test_labels)
```
